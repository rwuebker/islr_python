```python
# for vectors use numpy ndarrays
import numpy as np 
w = np.array([1,3,2,5])
print 'this is w: {}'.format(w)
x = np.array([1,6,2])
print 'now this is x: {}'.format(x)
y = np.array([1,4,3])
print 'this is y: {}'.format(y)
print 'this is length of y: {}'.format(len(y)) 
print 'this is x + y = {}'.format(x + y)
# no simple way in python to list all user defined variables
```

```python
# matricies
x = np.array([1,2,3,4]).reshape(2,2)
print 'this is our matrix: {}'.format(x)
print 'this is the square root of our matrix: {}'.format(np.sqrt(x)) # numpy sqrt function
import math # or you can import the math library
print 'this is the square root of 4: {}'.format(math.sqrt(4))
# math.sqrt won't work on arrays, only numpy can sqrt those
```

```python
# random numbers
# setting the seed will make the numbers predictable
np.random.seed(4)
print np.random.normal(0,1,5) # gives a list of 5 numbers sampled from a std normal dist (mean 0, std 1)
x = np.random.normal(0,1,10000)
print x.mean()
print x.std()
print x.var()
```

```python
# plotting
import matplotlib.pyplot as plt
import seaborn # adds jazz to matplotlib
%matplotlib inline
x = np.arange(50)
y = np.array(x*x)
z = np.array(x*x*x)
fig = plt.figure()
ax = fig.add_subplot(111) # 111 is shorthand for 1,1,1 which is the number of rows, columns and chart number
ax.plot(x,y)
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax1.plot(x,y)
ax1.set_title("squared")
ax2 = fig.add_subplot(122)
ax2.plot(x,z)
ax2.set_title('cubed')
fig.suptitle('Two Charts')
```

```python
# sequences use numpy linspace to include both lower and upper bounds and produce a certain number of evenly spaced nums
x = np.linspace(1,10, num = 5)
print x
# or numpy arange to up to but not including the upper bound
y = np.arange(1,10,2)
print y
```

```python
# 3d plotting
import numpy as np
import matplotlib.pyplot as plt
import seaborn
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
%matplotlib inline
from matplotlib.ticker import LinearLocator, FormatStrFormatter

fig = plt.figure()
ax = fig.gca(projection='3d')
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
ax.set_zlim(-1.01, 1.01)

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

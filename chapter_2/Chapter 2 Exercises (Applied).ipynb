```python
# make Elite a factor if > 50% are Top10perc
def top_fifty(x):
    if x > 50:
        return "yes"
    
    return "no"

college['Elite'] = college.Top10perc.apply(lambda my_var: "yes" if my_var > 50 else "no")
college['Elite2'] = ["yes" if x > 50 else "no" for x in college.Top10perc]
college['Elite3'] = college.Top10perc.apply(top_fifty)
```
